```python
# import needed libraries
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create data
data = {'Subfield': ['Field A', 'Field B', 'Field C'],
        '1': [1, 2, 3], '2': [3, 2, 7], '3': [1, 4, 7], '4': [2, 4, 9], '5': [8, 4, 12]}
  
# create dataframe
df = pd.DataFrame(data)
df = df.set_index('Subfield')

# for convertin a row to an array, use df.iloc[0].to_numpy(), for columns, use df.['column_name'].to_numpy() instead
field_a_array = df.iloc[0].to_numpy()
field_b_array = df.iloc[1].to_numpy()
field_c_array = df.iloc[2].to_numpy()
year_array = df.columns.values.astype(int)
```

```python
# specify dimensions of plot
width = 10
height = 6

# plot arrays
fig, ax = plt.subplots(figsize = [width, height], dpi = 100, facecolor = "white", edgecolor = "white")

# set title and axis titles
ax.set_xlabel('Year', size = 10, **{'fontname':'serif'})
ax.set_ylabel('Value', size = 10, **{'fontname':'serif'})
ax.set_title('Title', fontweight = 'bold', size = 12, loc='left', y=1.02, **{'fontname':'serif'})

# set xlim and ylim if needed
ax.set_xlim([0.5, 5.5])
ax.set_ylim([0, 14])

# adjust formatting of ticks
for tick in ax.get_xticklabels():
    tick.set_fontname('serif')
for tick in ax.get_yticklabels():
    tick.set_fontname('serif')

# add annotation if needed
plt.annotate(('Note: Some Annotation'), (0,0), (0, -50), xycoords = 'axes fraction', textcoords = 'offset points', size = 10,
             **{'fontname':'serif'})

# plot field a as a bar chart
ax.bar(year_array, field_a_array, color='blue', label ='Field A')

# plot field b as a line
ax.plot(year_array, field_b_array, color='red', linewidth=1, label ='Field B')

# plot field c as a scatterplot (black dots) with a blue regression line
#(change lowess=True, and ci+True if you want local regression smoother and confidence interval)
sns.regplot(x=year_array, y=field_c_array, scatter=True, ci=False, lowess=False, scatter_kws={'color': 'orange'},
            line_kws={'color': 'green', 'linewidth':1}, label ='Field C')

# add legend if it aids understanding
plt.legend()

plt.show()
```

```python
# save plot as png and pdf (keep dpi at 800 or higher for good resolution)

fig.savefig('filename.pdf', bbox_inches='tight', dpi=800)
fig.savefig('filename.png', bbox_inches='tight', dpi=800)
```
