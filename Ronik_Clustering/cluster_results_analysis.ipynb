```python
filenames = []
slopes = []
intercepts = []
r_squared_values = []
p_values = []
std_errs = []

for outfile in os.listdir(OUT_DIR):
    outfile = os.path.join(OUT_DIR, outfile)
    df = pd.read_csv(outfile)
    filenames.append(outfile)
    slope, intercept, r_value, p_value, std_err = stats.linregress(df["Optimal_Num_Clusters"],df["Year"] - 2000)
    slopes.append(slope)
    intercepts.append(intercept)
    r_squared_values.append(r_value ** 2)
    p_values.append(p_value)
    std_errs.append(std_err)


df = pd.DataFrame([row for row in zip(filenames, slopes, intercepts, r_squared_values, p_values, std_errs)],
                  columns=["File", "Slope", "Intercept", "R^2", "p-value", "std_err"])
```
